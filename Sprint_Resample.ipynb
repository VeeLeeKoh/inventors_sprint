```python
# Data Resampling
# Objective: Resample Inventors Sprint csv data on a monthly basis for data scrubbing

import pandas as pd
import numpy as np
```

```python
# read csv
consumption = pd.read_csv("Consumption.csv")
exports = pd.read_csv("Exports.csv")
imports = pd.read_csv("Imports.csv")
production = pd.read_csv("Production.csv")
refinery_production = pd.read_csv("Refinery_production.csv")
refinery_capacity = pd.read_csv("Refining_capacity.csv")
price = pd.read_csv("WTI_price.csv")
```

```python
production = production[["U.S. Field Production of Crude Oil (Thousand Barrels)","Date"]]
#consumption = consumption[["Weekly U.S. Product Supplied of Petroleum Products  (Thousand Barrels per Day)","Date"]]
#exports = exports[["Weekly U.S. Exports of Crude Oil and Petroleum Products  (Thousand Barrels per Day)","Date"]]
#imports = imports[["Weekly U.S. Imports of Crude Oil and Petroleum Products  (Thousand Barrels per Day)","Date"]]
#refinery_capacity.isna().sum()
```

```python
# dataframe 
dataList = [price,production,consumption,exports,imports,refinery_capacity,refinery_production]
resampledList = []
for df in dataList:
    df['Date'] = df['Date'].apply(pd.to_datetime)  # Convert date strings into datetime 
    df.set_index('Date',inplace=True)  # Set date as index
    resampledList.append(df.resample(rule = 'M').mean())  # Resample data on monthly average
```

```python
resampled_data = pd.concat(resampledList, sort=False, axis = 1)  # Concatenate all csv dataframes based on date
```

```python
resampled_data.to_csv("Sprint_Resampled Data.csv",index = True)  # Export as csv
```
