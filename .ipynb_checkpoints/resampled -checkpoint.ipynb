```python
import pandas as pd
import numpy as np
```

```python
# read csv
consumption = pd.read_csv("Consumption.csv")
exports = pd.read_csv("Exports.csv")
imports = pd.read_csv("Imports.csv")
production = pd.read_csv("Production.csv")
refinery_production = pd.read_csv("Refinery_production.csv")
refinery_capacity = pd.read_csv("Refining_capacity.csv")
price = pd.read_csv("WTI_price.csv")
```

```python
# dataframe 
dataList = [consumption,exports,imports,production,refinery_capacity,refinery_production,price]
resampled = []
for df in dataList:
    df['Date'] = df['Date'].apply(pd.to_datetime) 
    df.set_index('Date',inplace=True)
    resampled.append(df.resample(rule = 'M').mean())
```

```python
final = pd.concat(resampled, sort=True,axis = 1)
```

```python
final.to_csv("resampled.csv",index = True)
```

```python
final['Cushing, OK WTI Spot Price FOB (Dollars per Barrel)'].plot()
```
